# Clasification Model
#### Predict the wine 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from xgboost import XGBClassifier, plot_importance
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, f1_score, classification_report
import matplotlib.pyplot as plt
import pickle
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
%load_ext version_information
%matplotlib inline
```

```python
%version_information pandas, sklearn, matplotlib, xgboost
```

## 1.Load and prepare de data 

```python
train = pd.read_pickle('wine_train.pkl')
test = pd.read_pickle('wine_test.pkl')
X_train = train.drop(columns=['color'])
y_train = train[['color']]
X_test = test.drop(columns=['color'])
y_test = test[['color']]
```

## 2. Instance our models

```python
dt = DecisionTreeClassifier()
lr = LogisticRegression()
kn = KNeighborsClassifier()
svc = SVC()
rf = RandomForestClassifier()
xgb = XGBClassifier()
bg = BaggingClassifier()

#our scaler too
scaler = StandardScaler()
```

## 3. Create pipelines

```python
pipeline_dt = Pipeline(steps=[("scaler", scaler), ('dt', dt)])
pipeline_lr = Pipeline(steps=[('scaler', scaler), ('lr', lr)])
pipeline_kn = Pipeline(steps=[('scaler', scaler), ('kn', kn)])
pipeline_svc = Pipeline(steps=[('scaler', scaler), ('svc', svc)])
pipeline_rf = Pipeline(steps=[('scaler', scaler), ('rf', rf)])
pipeline_xgb = Pipeline(steps=[('scaler', scaler), ('xgb', xgb)])
pipeline_bg = Pipeline(steps=[('scaler', scaler), ('bg', bg)])
```

## 4. Grid parameters

```python
grid_dt = {"scaler__with_mean": [True, False],
           "scaler__with_std": [True, False],
           "dt__max_depth": [2,3,4,5,6],
           "dt__min_samples_split": [3,4,5,10],
           "dt__min_samples_leaf": [3,4,5,10],
           "dt__class_weight": [None, "balanced"]}

grid_lr = {"scaler__with_mean": [True, False],
           "scaler__with_std": [True, False],
           "lr__C": [0.01, 0.1, 0.5, 1.0, 2.0, 5.0],
           "lr__fit_intercept": [True, False],
           "lr__class_weight": [None, "balanced"]}

grid_kn = {"scaler__with_mean": [True, False],
            "scaler__with_std": [True, False],
            "kn__n_neighbors": [1,3,5,7,9,11],
            "kn__metric": ["minkowski", "euclidean"],
            "kn__p": [1,2,3,4]}

grid_svc = {"scaler__with_mean": [True, False],
            "scaler__with_std": [True, False],
            "svc__C": [0.01, 0.1, 0.5, 1.0, 2.0, 5.0],
            "svc__kernel": ["linear", "rbf"],
            "svc__degree": [2,3,4],
            "svc__class_weight": [None, "balanced"]}

grid_rf = {"scaler__with_mean": [True, False],
           "scaler__with_std": [True, False],
           "rf__n_estimators": [10, 30, 50, 100, 150, 200],
           "rf__max_depth": [3,4,5,6],
           "rf__min_samples_split": [3,4,5,10],
           "rf__min_samples_leaf": [3,4,5,10],
           "rf__class_weight": [None, "balanced"]}

grid_xgb = {"scaler__with_mean": [True, False],
            "scaler__with_std": [True, False],
            "xgb__learning_rate": [0.1, 0.01],
            "xgb__gamma" : [0.3, 0.5, 1,],
            "xgb__max_depth": [4, 7,],
            "xgb__n_estimators": [100, 250, 500, 1000]}

grid_bg = {"scaler__with_mean": [True, False],
           "scaler__with_std": [True, False],
           "bg__base_estimator":[lr],
           "bg__n_estimators": [30, 50, 70], 
           "bg__max_samples": [0.2, 0.4, 0.6], 
           "bg__max_features": [1,2,3,4]}
```

## 5. Grid Search instances

```python
gs_dt = GridSearchCV(estimator=pipeline_dt,
                     param_grid=grid_dt,
                     scoring="roc_auc",
                     cv=10,
                     n_jobs=-1)

gs_lr = GridSearchCV(estimator=pipeline_lr,
                     param_grid=grid_lr,
                     scoring="roc_auc",
                     cv=10,
                     n_jobs=-1)

gs_kn = GridSearchCV(estimator=pipeline_kn,
                      param_grid=grid_kn,
                      scoring="roc_auc",
                      cv=10,
                      n_jobs=-1)

gs_svc = GridSearchCV(estimator=pipeline_svc,
                      param_grid=grid_svc,
                      scoring="roc_auc",
                      cv=10,
                      n_jobs=-1)

gs_rf = GridSearchCV(estimator=pipeline_rf,
                     param_grid=grid_rf,
                     scoring="roc_auc",
                     cv=10,
                     n_jobs=-1)

gs_xgb = GridSearchCV(estimator=pipeline_xgb,
                     param_grid=grid_xgb,
                     scoring="roc_auc",
                     cv=10,
                     n_jobs=-1)

gs_bg = GridSearchCV(estimator=pipeline_bg,
                     param_grid=grid_bg,
                     scoring="roc_auc",
                     cv=10,
                     n_jobs=-1)
```

## 6. Train and test

```python
bag_of_models = {"Decision Tree": gs_dt,
                 "LR": gs_lr,
                 "KN": gs_kn,
                 "SVC": gs_svc,
                 "Random Forest": gs_rf,
                 "XGB": gs_xgb,
                 "GB": gs_bg}
```

```python
start = time.time()
for name, gs in bag_of_models.items():
    print("Doing Grid Search of {}".format(name))
    gs.fit(X_train, y_train)
    model = gs.best_estimator_
    model.fit(X_train, y_train)
    train_predict_ac = accuracy_score(model.predict(X_train), y_train)
    train_predict_roc = roc_auc_score(model.predict(X_train), y_train)
    print("Model accuracy in train {}".format(train_predict_ac))
    print("Model roc auc in train {}".format(train_predict_roc))
    test_predict_ac = accuracy_score(model.predict(X_test), y_test)
    test_predict_roc = roc_auc_score(model.predict(X_test), y_test)
    print('Model accuracy in test {}'.format(test_predict_ac))
    print('Model roc auc in test {}'.format(test_predict_roc))
    print("MODEL {} FINISHED".format(name))
end = time.time()
print("Time training model {} min".format(np.around((end - start)/60)))
```

All our models fit verry well, we can see that XGBoost have the best performance in train and test, so we are going to pick this model and train again with the best hyperparameters

```python
xgb_champion = XGBClassifier(base_score=0.5, booster='gbtree',
                               colsample_bylevel=1, colsample_bynode=1,
                               colsample_bytree=1, gamma=0.3, learning_rate=0.1,
                               max_delta_step=0, max_depth=7,
                               min_child_weight=1, missing=None,
                               n_estimators=250, n_jobs=1, nthread=None,
                               objective='binary:logistic', random_state=0,
                               reg_alpha=0, reg_lambda=1, scale_pos_weight=1,
                               seed=None, silent=None, subsample=1,
                               verbosity=1)


xgb_champion.fit(X_train, y_train)
print(classification_report(xgb_champion.predict(X_test), y_test))
print('Accuracy:', accuracy_score(xgb_champion.predict(X_test), y_test))
```

## 7 Interprete XGBoost

Below we have three plots of our XGB model
1. Weight: the number of times a feature is used to split te data across all threes
2. Cover: the number of times a feature is used to split the data across all trees weighted by the number of training data points that go through those splits.
3. Gain: the average training loss reduction gained when using a feature for splitting.

In these plots we can see clearly that there are two features creating noise in our model.
1. Quality: obviously the quality is not a good feature to measure the color of a wine. There are good/bad wines with different colors
2. Free sulfur dioxide: as we saw when prepropcesing the data, "total sulfur dioxide" and "free sulfur dioxide" are very correlated, so this variable is noise for our model

There are another features that is not clear if are good or bad for our model, for example, citric acid,fixed acidity, etc. And we will need a more deep analysis with other methods to know it, so to not extend this notebook too much we're going to leave it that way.

Here we can see that total sulfur dioxide, chlorides, sulphates, densitiy and ph are the most importan variables that affect to the color of the wine

```python
print(accuracy_score(xgb_champion.predict(X_train_clean), y_train))
print(accuracy_score(xgb_champion.predict(X_test_clean), y_test))
```

We obtain the same accuracy. This is because the models based in trees can handle thousand of features and obtain a good predictions without making a feature selection.
