# Regression Model

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
from xgboost import XGBRegressor, plot_importance
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import pickle
import time


import warnings
warnings.filterwarnings('ignore')
```

```python
%load_ext version_information
%version_information numpy, pandas, sklearn, xgboost
```

```python
train = pd.read_pickle('wine_train.pkl')
test = pd.read_pickle('wine_test.pkl')
X_train = train.drop(columns=['color', 'quality'])
y_train = train[['quality']]
X_test = test.drop(columns=['color', 'quality'])
y_test = test[['quality']]
```

```python
dt = DecisionTreeRegressor()
lr = LinearRegression()
kn = KNeighborsRegressor()
svr = SVR()
rf = RandomForestRegressor()
xgb = XGBRegressor()
bg = BaggingRegressor()

#our scaler too
scaler = StandardScaler()
```

```python
pipeline_dt = Pipeline(steps=[("scaler", scaler), ('dt', dt)])
pipeline_lr = Pipeline(steps=[('scaler', scaler), ('lr', lr)])
pipeline_kn = Pipeline(steps=[('scaler', scaler), ('kn', kn)])
pipeline_svr = Pipeline(steps=[('scaler', scaler), ('svr', svr)])
pipeline_rf = Pipeline(steps=[('scaler', scaler), ('rf', rf)])
pipeline_xgb = Pipeline(steps=[('scaler', scaler), ('xgb', xgb)])
pipeline_bg = Pipeline(steps=[('scaler', scaler), ('bg', bg)])
```

```python
grid_dt = {"scaler__with_mean": [True, False],
           "scaler__with_std": [True, False],
           "dt__max_depth": [2,3,4,5,6],
           "dt__min_samples_split": [3,4,5,10],
           "dt__min_samples_leaf": [3,4,5,10]}

grid_lr = {"scaler__with_mean": [True, False],
           "scaler__with_std": [True, False],
           "lr__fit_intercept": [True, False]}

grid_kn = {"scaler__with_mean": [True, False],
            "scaler__with_std": [True, False],
            "kn__n_neighbors": [1,3,5,7,9,11],
            "kn__metric": ["minkowski", "euclidean"],
            "kn__p": [1,2,3,4]}

grid_svr = {"scaler__with_mean": [True],
            "scaler__with_std": [True],
            "svr__C": [0.01, 0.1, 0.5, 1.0, 2.0, 5.0],
            "svr__kernel": ["linear", "rbf"],
            "svr__degree": [2,3,4]}

grid_rf = {"scaler__with_mean": [True, False],
           "scaler__with_std": [True, False],
           "rf__n_estimators": [10, 30, 50, 100, 150, 200],
           "rf__max_depth": [3,4,5,6],
           "rf__min_samples_split": [3,4,5,10],
           "rf__min_samples_leaf": [3,4,5,10]}

grid_xgb = {"scaler__with_mean": [True, False],
            "scaler__with_std": [True, False],
            "xgb__learning_rate": [0.1, 0.01],
            "xgb__gamma" : [0.3, 0.5, 1,],
            "xgb__max_depth": [4, 7,],
            "xgb__n_estimators": [100, 250, 500, 1000]}

grid_bg = {"scaler__with_mean": [True, False],
           "scaler__with_std": [True, False],
           "bg__base_estimator":[lr],
           "bg__n_estimators": [30, 50, 70], 
           "bg__max_samples": [0.2, 0.4, 0.6], 
           "bg__max_features": [1,2,3,4]}
```

```python
gs_dt = GridSearchCV(estimator=pipeline_dt,
                     param_grid=grid_dt,
                     scoring="neg_mean_squared_error",
                     cv=10,
                     n_jobs=-1)

gs_lr = GridSearchCV(estimator=pipeline_lr,
                     param_grid=grid_lr,
                     scoring="neg_mean_squared_error",
                     cv=10,
                     n_jobs=-1)

gs_kn = GridSearchCV(estimator=pipeline_kn,
                      param_grid=grid_kn,
                      scoring="neg_mean_squared_error",
                      cv=10,
                      n_jobs=-1)

gs_svr = GridSearchCV(estimator=pipeline_svr,
                      param_grid=grid_svr,
                      scoring="neg_mean_squared_error",
                      cv=10,
                      n_jobs=-1)

gs_rf = GridSearchCV(estimator=pipeline_rf,
                     param_grid=grid_rf,
                     scoring="neg_mean_squared_error",
                     cv=10,
                     n_jobs=-1)

gs_xgb = GridSearchCV(estimator=pipeline_xgb,
                     param_grid=grid_xgb,
                     scoring="neg_mean_squared_error",
                     cv=10,
                     n_jobs=-1)

gs_bg = GridSearchCV(estimator=pipeline_bg,
                     param_grid=grid_bg,
                     scoring="neg_mean_squared_error",
                     cv=10,
                     n_jobs=-1)
```

```python
bag_of_models = {"Decision Tree": gs_dt,
                 "LR": gs_lr,
                 "KN": gs_kn,
                 "SVR": gs_svr,
                 "Random Forest": gs_rf,
                 "XGB": gs_xgb,
                 "GB": gs_bg}
```

```python
start = time.time()
for name, gs in bag_of_models.items():
    print("Doing Grid Search of {}".format(name))
    gs.fit(X_train, y_train)
    model = gs.best_estimator_
    model.fit(X_train, y_train)
    train_predict_mse = mean_squared_error(model.predict(X_train), y_train)
    train_predict_mae = mean_absolute_error(model.predict(X_train), y_train)
    print("RMSE in train {}".format(np.sqrt(train_predict_mse)))
    print("MAE in train {}".format(train_predict_mae))
    test_predict_mse = mean_squared_error(model.predict(X_test), y_test)
    test_predict_mae = mean_absolute_error(model.predict(X_test), y_test)
    print('RMSE in test {}'.format(np.sqrt(test_predict_mse)))
    print('MAE in test {}'.format(test_predict_mae))
    print("MODEL {} FINISHED".format(name))
end = time.time()
print("Time training model {} min".format(np.around((end - start)/60)))
```

As in classification, XGB is the best model with the lowest RMSE.This model has gained a lot of fame in recent years and always tends to give good results. We can see that there is a little over fitting, in order to avoid it, we will change the parameters of our grid search result.

```python
bag_of_models['XGB'].best_estimator_
```

```python
xgb_champion = XGBRegressor(base_score=0.5, booster='gbtree',
                              colsample_bylevel=0.3, colsample_bynode=1,
                              colsample_bytree=1, gamma=0,
                              importance_type='gain', learning_rate=0.01,
                              max_delta_step=0, max_depth=7, min_child_weight=1,
                              missing=None, n_estimators=650, n_jobs=-1,
                              nthread=None, objective='reg:linear',
                              random_state=0, reg_alpha=0, reg_lambda=1,
                              scale_pos_weight=1, seed=None, silent=None,
                              subsample=0.50, verbosity=1)


xgb_champion.fit(X_train, y_train)
print(np.sqrt(mean_squared_error(xgb_champion.predict(X_train), y_train)))
print(np.sqrt(mean_squared_error(xgb_champion.predict(X_test), y_test)))
```
