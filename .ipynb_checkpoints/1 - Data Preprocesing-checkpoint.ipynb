**So now we are going to:**
1. Drop duplicates
2. Binarize our wine class
3. Separate in train and test: we need to do that now because we can't clean the outliers in the test set
4. Clean outliers

```python
# 1- Temoving duplicates
print('data before removing duplicates {}'.format(wines.shape))
wine_no_dup = wines.drop_duplicates().copy()
print('data after removing duplicates {}'.format(wine_no_dup.shape))
```

```python
# 2- Binarize our wine class
binarizer = LabelEncoder()
wine_no_dup['color'] = binarizer.fit_transform(wine_no_dup['color'])
print('The number 0 is the {} wine and the number 1 is the {} wine'.format(binarizer.classes_[0],binarizer.classes_[1]))
```

```python
# 3. Separate our train and test
train, test = train_test_split(wine_no_dup, train_size=0.85, test_size=0.15)
```

```python
# 4. Clean outliers
train = train.reset_index(drop=True)

def outlier_cleaner(df, sigma=3):
    '''
    A function to clear all the outliers of a dataframe
    
    :param df: The dataframe to clean outliers
    :param sigma: The z_score threshold to remove a value, by default 3
    :return: the dataframe with the outliers removed
    '''
    for col in df.columns:
        column_data = df[col]
        mean = np.mean(column_data)
        std = np.std(column_data)
        

        outliers = []
        counter = 0
        
        for data in column_data:
            z_score = (data -mean)/std
            if np.abs(z_score) >= sigma:
                outliers.append(counter)
            counter += 1
            
        df.drop(df.index[outliers], inplace=True)

    df.reset_index(drop=True, inplace=True)
    return df
    
```

```python
#use our function
train = outlier_cleaner(train)
train
```

```python
#Will run again pandas profiling to see the quality of our data
ProfileReport(train)
```
